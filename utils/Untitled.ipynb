```python
import pandas as pd
```

```python
df = pd.read_csv("construction_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df["cost_per_sqft_est"] = df["material_price_index"] * df["material_quality"]
```

```python
df["labor_intensity"] = df["labor_cost"] / df["area"]
```

```python
df["risk_index"] = (df["weather_index"] +(df["soil_type"] / 3) +(df["project_complexity"] / 3))
```

```python
df["efficiency"] = (df["contractor_experience"] *df["equipment_availability"])
```

```python
df.info()
```

```python
df.to_csv("data.csv")
```
