```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("data2.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
cat_cols = ["material", "location", "quality"]
```

```python
encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoders[col] = le
```

```python
X = df[["area", "floors", "material", "location", "quality"]]
```

```python
y_labor = df["labor_cost"]
```

```python
y_material = df["material_cost"]
```

```python
X_train, X_test, y_labor_train, y_labor_test, y_material_train, y_material_test = train_test_split(
    X, y_labor, y_material, test_size=0.2, random_state=42
)
```

```python
labor_model = RandomForestRegressor(
    n_estimators=200,
    max_depth=12,
    min_samples_split=5,
    random_state=42,
    n_jobs=-1
)

material_model = RandomForestRegressor(
    n_estimators=200,
    max_depth=12,
    min_samples_split=5,
    random_state=42,
    n_jobs=-1
)

labor_model.fit(X_train, y_labor_train)
material_model.fit(X_train, y_material_train)
```

```python
labor_pred = labor_model.predict(X_test)
```

```python
material_pred = material_model.predict(X_test)
```

```python
total_pred = labor_pred + material_pred
```

```python
print("Labor MAE:", mean_absolute_error(y_labor_test, labor_pred))
print("Labor R2:", r2_score(y_labor_test, labor_pred))

print("Material MAE:", mean_absolute_error(y_material_test, material_pred))
print("Material R2:", r2_score(y_material_test, material_pred))
```

```python
import joblib

joblib.dump(labor_model, "labor_model.pkl")
joblib.dump(material_model, "material_model.pkl")
joblib.dump(encoders, "encoders.pkl")
```
